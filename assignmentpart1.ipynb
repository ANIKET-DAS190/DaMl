```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df1['destination_encoded'] = le.fit_transform(df1['destination'])
df1['passanger_encoded'] = le.fit_transform(df1['passanger'])
df1['weather_encoded'] = le.fit_transform(df1['weather'])
df1['time_encoded'] = le.fit_transform(df1['time'])
df1['coupon_encoded'] = le.fit_transform(df1['coupon'])
df1['expiration_encoded'] = le.fit_transform(df1['expiration'])
df1['gender_encoded'] = le.fit_transform(df1['gender'])
df1['age_encoded'] = oe.fit_transform(df1[['age']])
df1['maritalStatus_encoded'] = le.fit_transform(df1['maritalStatus'])
df1['education_encoded'] = le.fit_transform(df1['education'])
df1['occupation_encoded'] = le.fit_transform(df1['occupation'])
df1['Bar_encoded'] = oe.fit_transform(df1[['Bar']])
df1['CoffeeHouse_encoded'] = oe.fit_transform(df1[['CoffeeHouse']])
df1['CarryAway_encoded'] = oe.fit_transform(df1[['CarryAway']])
df1['RestaurantLessThan20_encoded'] = oe.fit_transform(df1[['RestaurantLessThan20']])
df1['Restaurant20To50_encoded'] = oe.fit_transform(df1[['Restaurant20To50']])
```

```python
df1.columns
```

```python
X = df1[['income_encoded', 'temperature_scaled',
       'destination_encoded', 'passanger_encoded', 'weather_encoded',
       'time_encoded', 'coupon_encoded', 'expiration_encoded',
       'gender_encoded', 'age_encoded', 'maritalStatus_encoded',
       'education_encoded', 'occupation_encoded', 'Bar_encoded',
       'CoffeeHouse_encoded', 'CarryAway_encoded',
       'RestaurantLessThan20_encoded', 'Restaurant20To50_encoded','toCoupon_GEQ5min',
       'toCoupon_GEQ15min', 'toCoupon_GEQ25min', 'direction_same',
       'direction_opp', 'has_children']]
Y = df1['Y']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,Y_train)
```

```python
Y_pred1 = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy_score(Y_test, Y_pred1)
```

```python
f1_score(Y_test, Y_pred1)
```

```python
precision_score(Y_test, Y_pred1)
```

```python
recall_score(Y_test, Y_pred1)
```
